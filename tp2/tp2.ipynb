```python
import pandas as pd
from scipy.stats import t
import numpy as np
import scipy.stats as stats

espessuras = np.array([30, 30, 30, 30, 31, 32, 32, 32, 32, 33, 33, 34, 34, 34, 35])

# Cálculo da média e desvio padrão da amostra
media_amostral = np.mean(espessuras)
desvio_padrao_amostral = np.std(espessuras, ddof=1)  # ddof=1 para amostra
tamanho_amostra = len(espessuras)

# Hipótese nula
mu_0 = 32.5

# Cálculo do valor t
valor_t = (media_amostral - mu_0) / (desvio_padrao_amostral / np.sqrt(tamanho_amostra))

# Cálculo do p-valor para duas caudas
p_valor = 2 * t.sf(np.abs(valor_t), df=tamanho_amostra - 1)

print("Media Amostral", round(media_amostral, 4))
print("Desvio Padrão Amostral", round(desvio_padrao_amostral, 4))
print("Valor Absoluto", round(valor_t, 4))
print("P-Valor para as duas caudas", round(p_valor, 4))
```

```python
statistic, p_value = stats.ttest_1samp(espessuras, popmean=mu_0, alternative='two-sided')
print("Valor Absoluto", round(statistic, 4))
print("P-Valor para as duas caudas", round(p_value, 4))
```

```python
# Dados da amostra
entrega_dias = [
    5, 4, 4, 5, 5, 5, 6, 5, 4, 4, 3, 4, 4, 5, 5, 7, 6, 5,
    6, 4, 6, 5, 5, 6, 6, 6, 4, 4, 5, 5, 5, 3, 6, 3, 6, 5
]

# Realizando o teste t de uma amostra
t_stat, p_value = stats.ttest_1samp(entrega_dias, popmean=5)

# Ajustando o p-valor para um teste unicaudal (a função retorna o p-valor para um teste bilateral)
p_value_unicaudal = p_value / 2

t_stat, p_value_unicaudal
```

Para ultrapassar o nível de significância e, assim, fornecer evidência estatística suficiente para rejeitar a hipótese nula em um teste com nível de significância de 1% (\(\alpha = 0.01\)), o p-valor precisaria ser menor que 0.01. Em termos práticos, isso significa que a probabilidade de observar os dados da amostra, ou algo mais extremo, sob a suposição de que a hipótese nula é verdadeira, teria de ser inferior a 1%.

Um p-valor menor que 0.01 indicaria que os resultados são altamente improváveis sob a hipótese nula, sugerindo que há uma diferença significativa entre a média da amostra e a média populacional proposta pela hipótese nula (\(\mu = 5\) dias, neste caso). Essa seria a evidência necessária para rejeitar a hipótese nula e aceitar a hipótese alternativa de que o tempo médio de entrega é superior ao anunciado pela loja online, com um nível de confiança de 99% (ou seja, cometendo um erro do Tipo I com probabilidade de apenas 1%).

```python
velocidades_com_fio= [2300,2000,1800,2000,2400,2200,2000,1800,1900,2100,2200,2400]
velocidades_sem_fio= [2400,2200,1800,1900,1800,1900,2100,2050,2200,2000,1900,2000]
```

Para determinar se os grupos de ratos de computador "Com fio" e "Sem fio" são independentes ou dependentes, precisamos considerar a natureza do experimento e como os dados foram coletados:

1. **Grupos Independentes:** Os grupos seriam considerados independentes se os ratos selecionados para cada grupo fossem diferentes e não relacionados entre si. Isso significa que a seleção de um rato para um grupo não afetaria a seleção de ratos para o outro grupo. Em contextos experimentais, grupos independentes geralmente surgem de amostras aleatórias onde os participantes são distribuídos aleatoriamente entre os grupos de tratamento.

2. **Grupos Dependentes:** Os grupos seriam considerados dependentes (ou emparelhados) se os mesmos ratos fossem testados em ambos os grupos (o que não é o caso aqui) ou se houvesse uma correspondência direta entre os membros dos dois grupos (por exemplo, ratos emparelhados por peso, idade ou alguma outra característica relevante antes do teste). Outra situação comum de dependência ocorre em medidas repetidas, onde o mesmo sujeito é testado sob diferentes condições.

Dado o contexto fornecido, parece que temos dois grupos distintos de ratos de computador: um grupo com ratos "Com fio" e outro com ratos "Sem fio". Não há indicação de que os mesmos ratos foram testados em ambos os grupos ou que há uma correspondência direta entre os ratos dos dois grupos que justificaria tratá-los como emparelhados ou dependentes.

**Conclusão:** Baseado nas informações fornecidas, os grupos de ratos de computador "Com fio" e "Sem fio" parecem ser **independentes**, pois não há evidências de que os ratos em um grupo sejam diretamente relacionados ou correspondam aos ratos no outro grupo. A independência dos grupos é uma característica importante para determinar o tipo de teste estatístico a ser aplicado para comparar as médias dos dois grupos (por exemplo, um teste t para amostras independentes).

```python
# Teste t para duas amostras independentes
t_stat, p_value = stats.ttest_ind(velocidades_com_fio, velocidades_sem_fio,equal_var=True)

t_stat, p_value
```

Para testar a hipótese de igualdade das velocidades médias entre ratos de computador com e sem fio, assumindo um nível de significância de 1% e que a velocidade é uma variável aleatória normal, usaremos o teste t para duas amostras independentes. Este teste é apropriado dado que determinamos que os dois grupos são independentes.

As hipóteses são formuladas da seguinte maneira:
- **Hipótese Nula (\(H_0\))**: A média das velocidades dos ratos com fio é igual à média das velocidades dos ratos sem fio, ou seja, \(\mu_{\text{com fio}} = \mu_{\text{sem fio}}\).
- **Hipótese Alternativa (\(H_1\))**: As médias das velocidades diferem entre os dois grupos, ou seja, \(\mu_{\text{com fio}} \neq \mu_{\text{sem fio}}\).

Vamos calcular a estatística de teste t e o p-valor correspondente usando os dados fornecidos.

A estatística de teste t calculada é aproximadamente 1.63, e o p-valor correspondente é aproximadamente 0.117. 

Com um nível de significância de 1% (\(\alpha = 0.01\)), comparamos o p-valor obtido com o nível de significância. Visto que o p-valor (0.117) é maior do que 0.01, não rejeitamos a hipótese nula (\(H_0\)). Isso indica que, com base nos dados analisados e assumindo um nível de significância de 1%, não há evidências estatísticas suficientes para afirmar que existe uma diferença significativa nas velocidades médias entre os ratos de computador com fio e sem fio.

Portanto, concluímos que, ao nível de confiança de 99%, as velocidades médias para ratos de computador com e sem fio são estatisticamente iguais, segundo os dados analisados.
