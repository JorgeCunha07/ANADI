```python
import numpy as np
from scipy.stats import norm

p0 = 0.45  # Proporção esperada na população em geral
n = 25  # Número total de oficiais
X = 13  # Número de oficiais com resultado acima de 50 pontos
p_hat = X / n  # Proporção observada no grupo de oficiais

# Cálculo da estatística do teste Z
Z = (p_hat - p0) / np.sqrt((p0 * (1 - p0)) / n)

# Cálculo do p-valor para um teste unilateral à direita
p_value = 1 - norm.cdf(Z)

Z, p_value
```

A estatística de teste Z calculada é aproximadamente 0.704, e o p-valor associado é aproximadamente 0.241.

A hipótese alternativa é que há uma preferência pela cor vermelha, o que implica que a probabilidade de escolher o balde vermelho é maior do que 50%.

```python
from statsmodels.stats.proportion import proportions_ztest

# Dados: 700 de 2000 eleitores pretendem votar no partido P
count = 700  # número de sucessos (votos no partido P)
nobs = 2000  # número total de observações (eleitores na amostra)
value = 0.30  # proporção de referência sob H0

# Realizando o teste Z para uma proporção
stat, pval = proportions_ztest(count, nobs, value, alternative='larger')

print(f"Z statistic: {stat}, p-value: {pval}")
```

```python
import pandas as pd
from scipy.stats import chisquare

# Carregar os dados
dadosInfo = pd.read_csv("Ex5.csv", sep=';')

# Calcular as frequências observadas de cada resultado (1 a 6)
preferencias_1 = sum(dadosInfo['Resultado'] == 1)
preferencias_2 = sum(dadosInfo['Resultado'] == 2)
preferencias_3 = sum(dadosInfo['Resultado'] == 3)
preferencias_4 = sum(dadosInfo['Resultado'] == 4)
preferencias_5 = sum(dadosInfo['Resultado'] == 5)
preferencias_6 = sum(dadosInfo['Resultado'] == 6)

# Lista das frequências observadas
frequencias_observadas = [
    preferencias_1, preferencias_2, preferencias_3,
    preferencias_4, preferencias_5, preferencias_6
]

# Frequências esperadas para um dado não viciado (cada face esperada para sair igualmente em 60 lançamentos)
frequencias_esperadas = [10] * 6  # 60 lançamentos / 6 faces = 10 vezes cada face

# Realizar o teste do qui-quadrado
chi2_stat, p_value = chisquare(frequencias_observadas, f_exp=frequencias_esperadas)

print(f"Qui-Quadrado: {chi2_stat}, P-valor: {p_value}")
```

### Interpretação do Resultado:
- Se o **p-valor** for **menor que 0.05**, há evidências suficientes para rejeitar a hipótese nula, indicando que o dado pode estar viciado.
- Se o **p-valor** for **maior que 0.05**, não há evidências suficientes para rejeitar a hipótese nula, indicando que não há diferença significativa entre as frequências observadas e as esperadas, e o dado pode ser considerado não viciado.

Interpretação do Teste de Shapiro-Wilk:
Estatística de Teste: O valor da estatística de teste (0.826) nos dá uma medida de quão bem os dados se ajustam a uma distribuição normal. Valores mais próximos de 1 indicam um melhor ajuste à normalidade.

Conclusão:
Com um p-valor de 0.041, temos evidências suficientes para rejeitar a hipótese nula de que os dados seguem uma distribuição normal ao nível de significância de 5%. Isso sugere que os tempos de execução do algoritmo, com base nesta amostra de oito computadores, não seguem uma distribuição normal.

Implicações para Testes Subsequentes:
A rejeição da normalidade implica que testes que assumem a normalidade dos dados (como o teste t para uma amostra) podem não ser os mais apropriados. No entanto, com uma amostra pequena como esta (n=8), o teste de Shapiro-Wilk pode ser bastante sensível a desvios de normalidade. Para a questão do engenheiro informático sobre se o tempo de execução ultrapassa os 25 segundos, poderíamos considerar usar testes não paramétricos, como o teste de sinais ou o teste de Wilcoxon para uma amostra, que não assumem normalidade dos dados. Se ainda quisermos prosseguir com um teste que assume normalidade, precisamos ter em mente a limitação indicada pelo teste de Shapiro-Wilk.

```python
from scipy.stats import kstest, norm

# Calculando a média e o desvio padrão dos tempos de execução
media = norm.mean(tempos_execucao)
desvio_padrao = norm.std(tempos_execucao)

# Realizando o teste de Lilliefors (aproximado pelo teste KS com a distribuição normal ajustada)
lilliefors_stat, lilliefors_p = kstest(tempos_execucao, 'norm', args=(media, desvio_padrao))

lilliefors_stat, lilliefors_p
```

O teste de Kolmogorov-Smirnov (uma aproximação para o teste de Lilliefors neste contexto, ajustando para a média e desvio padrão da amostra) retornou uma estatística de teste de aproximadamente 0.397 e um p-valor de aproximadamente 0.086.

### Interpretação do Teste de Lilliefors (Aproximação pelo Teste KS):

- **Estatística de Teste**: O valor da estatística de teste (0.397) indica a máxima distância entre a função de distribuição acumulada (FDA) dos dados observados e a FDA esperada de uma distribuição normal com a mesma média e desvio padrão que os dados observados.
  
- **P-valor**: O p-valor de aproximadamente 0.086 é maior do que o nível de significância de 0.05 (\(\alpha = 0.05\)). Isso significa que, ao contrário do teste de Shapiro-Wilk, não temos evidências suficientes para rejeitar a hipótese nula de que os dados seguem uma distribuição normal ao nível de significância de 5%.

### Conclusão:

Com base no teste de Lilliefors (via aproximação pelo teste de Kolmogorov-Smirnov), os dados não mostram evidências significativas de desvio da normalidade ao nível de significância de 5%. Isso contrasta com o resultado do teste de Shapiro-Wilk, sugerindo que a decisão sobre a normalidade dos dados pode depender do teste específico utilizado e do nível de significância.

### Considerações:

A diferença nos resultados entre o teste de Shapiro-Wilk e a aproximação de Lilliefors pelo teste KS pode decorrer da sensibilidade de cada teste a diferentes aspectos dos dados (por exemplo, caudas da distribuição versus o ajuste geral à distribuição normal). Em prática, ao avaliar a normalidade, pode ser útil considerar os resultados de vários testes, bem como o tamanho da amostra e a importância prática de desvios da normalidade para a análise subsequente.

- **Skewness = 0**: Representaria uma distribuição perfeitamente simétrica.
