```python
#Exercicio A
import pandas as pd


dados = pd.read_excel("PORDATA_Dias_sem_chuva.xlsx", skiprows=7, nrows=9)
dados = dados.iloc[:, :10]
# Exercicio B
dados.columns.values[0] = "Ano"
dados.head(n=10)

```

```python
import matplotlib.pyplot as plt

# Criar o boxplot, excluindo a primeira coluna (ano)
ax = dados.iloc[:, 1:].boxplot(figsize=(12, 8), rot=45,grid=False)  # .iloc[:, 1:] exclui a primeira coluna

# Adicionar títulos e rótulos
ax.set_title('Dias Sem Chuva por Distrito')
ax.set_ylabel('Dias Sem Chuva')
ax.set_xlabel('Distrito')

# Mostrar o gráfico
plt.show()
```
